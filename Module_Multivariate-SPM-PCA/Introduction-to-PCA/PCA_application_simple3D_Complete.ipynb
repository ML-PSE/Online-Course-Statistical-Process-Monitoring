### *Topic*:  PCA Implementation using Sklearn on a simple 3D dataset

```python
# import required packages
import numpy as np, pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
np.set_printoptions(precision=3, suppress=True)
```

```python
# ==================================================================================
#                        Read and explore dataset
# ==================================================================================
```

```python
# fetch data and explore
df = pd.read_csv('3D_numericalDataset.csv')

print("\n First few rows:")
print(df.head())

print("\n \n Basic statistics:")
print(df.describe())
```

```python
# plot data
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(df['x'], df['y'], df['z'], c='brown', marker='o', alpha=0.6, s=25)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Points Loaded from CSV')

plt.show()
```

```python
# ==================================================================================
#                        Fit PCA model 
# ==================================================================================
```

```python
# convert Pandas dataframe to Numpy array
data = df.values # can continue working with dataframe as well
print('Numpy array:\n', data) 
```

```python
# scale data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
           
# fit PCA with 1 component
pca = PCA(n_components=1)
pca.fit(data_scaled)
score_1D = pca.transform(data_scaled) # could also do score_1D = pca.fit_transform(data_scaled)

print("Transformed data in 1D space:")
print(score_1D)
```

```python
# ==================================================================================
#              Reconstruct 3D dataset from the transformed 1D scores
# ==================================================================================
```

```python
# reconstruct from PCA
data_scaled_reconstructed = pca.inverse_transform(score_1D)
print('Reconstructed Scaled Numpy array:\n', data_scaled_reconstructed) 
```

```python
# scale back to original scale
data_reconstructed = scaler.inverse_transform(data_scaled_reconstructed)
print('Reconstructed Numpy array:\n', data_reconstructed) 
```

```python
# plot original and reconstructed data-points
#%matplotlib qt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(data[:,0], data[:,1], data[:,2], c='brown', marker='o', alpha=0.6, s=25, label='original')
ax.scatter(data_reconstructed[:,0], data_reconstructed[:,1], data_reconstructed[:,2], c='black', marker='o', alpha=0.3, s=15, label='reconstructed')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Points Loaded from CSV')

plt.legend()
plt.show()
```

>#### using PCA, we are able to remove the noisy component from data and 'extract' out the 1D line along which the observation samples lie!

```python
# ==================================================================================
#              Error or Residual matrix
# ==================================================================================
```

```python
errorMatrix = data - data_reconstructed
print('Error Matrix: \n', errorMatrix)
```

```python
# ==================================================================================
#                   Model inference
# ==================================================================================
```

```python
# generate a test sample
x = 6
y = x + np.random.normal(0, 2*0.1)
z = 20 - x - y + 5

# Stack coordinates into a single array
testSample = np.array((x,y,z))[None,:]
print(testSample)
```

```python
# plot training and test samples
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(data[:,0], data[:,1], data[:,2], c='brown', marker='o', alpha=0.6, s=25, label='training')
ax.scatter(testSample[:,0], testSample[:,1], testSample[:,2], c='red', marker='*', alpha=0.9, s=100, label='test')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('training and test samples')

plt.legend()
plt.show()
```

```python
# scale test sample
testSample_scaled = scaler.transform(testSample)
           
# apply fitted PCA model
testSample_score_1D = pca.transform(testSample_scaled)

# reconstruct 
testSample_scaled_reconstructed = pca.inverse_transform(testSample_score_1D)
testSample_reconstructed = scaler.inverse_transform(testSample_scaled_reconstructed)

print(testSample_reconstructed)
```

```python
# plot original and reconstructed data-points
# %matplotlib qt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(data[:,0], data[:,1], data[:,2], c='brown', marker='o', alpha=0.6, s=25, label='training')
ax.scatter(data_reconstructed[:,0], data_reconstructed[:,1], data_reconstructed[:,2], c='black', marker='o', alpha=0.3, s=15, label='reconstructed training')
ax.scatter(testSample[:,0], testSample[:,1], testSample[:,2], c='red', marker='*', alpha=0.9, s=100, label='test')
ax.scatter(testSample_reconstructed[:,0], testSample_reconstructed[:,1], testSample_reconstructed[:,2], c='black', marker='*', alpha=0.9, s=100, label='reconstructed test')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Points Loaded from CSV')

plt.legend()
plt.show()
```
