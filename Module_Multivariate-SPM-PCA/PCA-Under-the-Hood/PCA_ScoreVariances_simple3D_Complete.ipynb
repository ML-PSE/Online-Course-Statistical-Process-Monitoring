### *Topic*:  Variances captured by PCs

```python
# import required packages
import numpy as np, pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# ==================================================================================
#                        Read and explore dataset
# ==================================================================================
```

```python
# fetch data and explore
df = pd.read_csv('3D_numericalDataset.csv')

print("\n First few rows:")
print(df.head())

print("\n \n Basic statistics:")
print(df.describe())
```

```python
# plot data
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(df['x'], df['y'], df['z'], c='brown', marker='o', alpha=0.6, s=25)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Points Loaded from CSV')

plt.show()
```

```python
# ==================================================================================
#                        Fit PCA model 
# ==================================================================================
```

```python
# convert Pandas dataframe to Numpy array
data = df.values # can continue working with dataframe as well
print('Numpy array:\n', data) 
```

```python
# scale data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
           
# fit PCA (retaining all PCs by default)
pca = PCA()
pca.fit(data_scaled)
```

```python
# get the eigenvalues or the variances of the different PC Scores
variances_captured = pca.explained_variance_
print('variances of the PC scores: ', variances_captured)
```

##### Note that the variances are in decreasing order. The magnitude of the variance provides clues regarding the 'importance' of a principal component.
